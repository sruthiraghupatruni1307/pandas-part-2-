## concatination of data frames

```python
d1=pd.DataFrame({'clg':['spmvv','gist','aec','alts'],
                'temp':[35,47,68,33],
                'hum':[42,56,87,46]})
d1
```

```python
d2=pd.DataFrame({'clg':['mvgr','brau','au','jntu'],
                'temp':[65,25,68,23],
                'hum':[46,16,57,48]})
d2
```

```python
import pandas as pd
d3=pd.concat((d1,d2))
d3
```

```python
import pandas as pd
d3=pd.concat((d1,d2),ignore_index=True)
d3
```

```python
df1=pd.DataFrame({'lkey': ['foo', 'bar', 'baz', 'foo'],
                     'value': [1, 2, 3, 5]})
df2 = pd.DataFrame({'rkey': ['foo', 'bar', 'baz', 'foo'],
                    'value': [5, 6, 7, 8]})
df1
```

```python
df2
```

```python
df3=pd.merge(df1,df2,indicator=True)
df3
```

```python
df3=pd.merge(df1,df2,how='left',indicator=True)
df3
```
